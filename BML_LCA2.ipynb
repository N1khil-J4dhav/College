```python
# Step 1: Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Step 2: Load the Adult dataset
url_train = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
url_test = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.test'
```

```python
# Column names from the dataset description
columns = [
    "age", "workclass", "fnlwgt", "education", "education-num",
    "marital-status", "occupation", "relationship", "race", "sex",
    "capital-gain", "capital-loss", "hours-per-week", "native-country", "income"
]
```

```python
# Load train and test datasets
train_data = pd.read_csv(url_train, names=columns, sep=r',\s', engine='python', na_values='?')  # Use raw string r',\s'
test_data = pd.read_csv(url_test, names=columns, sep=r',\s', engine='python', skiprows=1, na_values='?')  # Use raw string r',\s'
```

```python
# Step 3: Preprocess the data

# Combine train and test for consistent encoding and then split again later
data = pd.concat([train_data, test_data], ignore_index=True)

# Drop rows with missing values
data = data.dropna()

# Remove trailing '.' from income column in test data
data['income'] = data['income'].apply(lambda x: x.replace('.', '') if isinstance(x, str) else x)

# Encode categorical variables
cat_columns = data.select_dtypes(include=['object']).columns.tolist()
cat_columns.remove('income')  # exclude target variable

label_encoders = {}
for col in cat_columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le

# Encode target variable
target_le = LabelEncoder()
data['income'] = target_le.fit_transform(data['income'])
```

```python
# Step 4: Split data back into train and test sets
train_len = len(train_data) - train_data['workclass'].isna().sum()  # Number of rows after dropna in train set

train_set = data[:train_len]
test_set = data[train_len:]

X_train = train_set.drop('income', axis=1)
y_train = train_set['income']

X_test = test_set.drop('income', axis=1)
y_test = test_set['income']
```

```python
# Step 5: Feature scaling for KNN
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Step 6: Train KNN classifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
# Step 7: Predict and calculate accuracy
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"KNN classifier accuracy on Adult dataset test set: {accuracy:.4f}")
```
